```python
X_train_flattend=X_train.reshape(len(X_train), X_train.shape[1]*X_train.shape[1])
X_test_flattend=X_test.reshape(len(X_test), X_test.shape[1]*X_test.shape[1])

print(X_train_flattend.shape)
print(X_test_flattend.shape)
```

Let's scale data and re-run

```python
X_train=X_train/255
X_test=X_test/255
X_train_flattend=X_train.reshape(len(X_train), X_train.shape[1]*X_train.shape[1])
X_test_flattend=X_test.reshape(len(X_test), X_test.shape[1]*X_test.shape[1])
```

```python
model=keras.Sequential([keras.layers.Dense(10,input_shape=(784,), activation='sigmoid')])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train_flattend, y_train, epochs=50)
```

```python
model.evaluate(X_test_flattend, y_test)
```

Let's add a hidden layer to improve the solution

```python
model=keras.Sequential([
    keras.layers.Dense(50,input_shape=(784,), activation='relu'), # 100 ( sgould be less than input) is the number of nueron in the hidden layer and 10 is the number of output
    keras.layers.Dense(10, activation='sigmoid')
    ])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train_flattend, y_train, epochs=50)
```

```python
model.evaluate(X_test_flattend, y_test)
```

```python
history = model.fit(X_train_flattend, y_train, epochs=50, validation_split=0.30)
training_loss = history.history['loss']
validation_loss = history.history['val_loss']
# plot the accuracy and loss graphs for train and test data

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
ax1.plot(history.history['loss'], label='train loss')
ax1.plot(history.history['val_loss'], label='test loss')

ax2.plot(history.history['accuracy'], label='train accuracy')
ax2.plot(history.history['val_accuracy'], label='test accuracy')

ax1.legend()
ax2.legend()
plt.show()
```

You see improvement. Let's add a hidden layer

```python
model=keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,), activation='relu'), # 100 ( sgould be less than input) is the number of nueron in the hidden layer and 10 is the number of output
    keras.layers.Dense(100,activation='relu'),   
    keras.layers.Dense(10, activation='sigmoid')
    ])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train_flattend, y_train, epochs=50)
```

You see that adding a hidden layer does not improve the accuracy. Let's use flatten function of keras instead of reshaping. 

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
    ])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=50)
```

```python
model.evaluate(X_test, y_test)
```

```python
history = model.fit(X_train_flattend, y_train, epochs=50, validation_split=0.30)
training_loss = history.history['loss']
validation_loss = history.history['val_loss']
# plot the accuracy and loss graphs for train and test data

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
ax1.plot(history.history['loss'], label='train loss')
ax1.plot(history.history['val_loss'], label='test loss')

ax2.plot(history.history['accuracy'], label='train accuracy')
ax2.plot(history.history['val_accuracy'], label='test accuracy')

ax1.legend()
ax2.legend()
plt.show()
```

***Post Training quantization***
Here we use post training Quantization to reduce the size of saved model

```python
converter = tf.lite.TFLiteConverter.from_saved_model("./saved_model")
tflite_model=converter.convert()
```

So far we used tflite to redice the saved model. Now we use quatization method to further decrease the saved model size.

```python
converter.optimizations=[tf.lite.Optimize.DEFAULT]
tflit_quant_model=converter.convert()
```

```python
len(tflite_model), len(tflit_quant_model)
```

You see huge reduction - nearly 1/4 (numbers show numbers of bytes). 
You can save models.

```python
with open('tflite_model', 'wb') as f: # since tflite_model is binary we use wb
    f.write(tflite_model)
```

```python
with open('tflit_quant_model', 'wb') as f:
    f.write(tflit_quant_model)
```

***Quantization aware training
Here you can get better accuracy

```python
import tensorflow_model_optimization as tfopt
quantize_model = tfopt.quantization.keras.quantize_model
q_aware_model=quantize_model(model)
q_aware_model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
q_aware_model.summary()
```

```python
q_aware_model.fit(X_train, y_train, epochs=1)
```

```python
q_aware_model.evaluate(X_test, y_test)
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(q_aware_model)
converter.optimizations=[tf.lite.Optimize.DEFAULT]
tflite_quant_model=converter.convert()
```

```python
len(tflite_quant_model)
```

```python
with open("tflite_qaware_model.tflite", 'wb') as f:
    f.write(tflite_quant_model)
```

the main benefit of this method is that the training accuracy is much higher than when you use Post Training quantization
